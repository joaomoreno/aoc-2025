```python
def parse(input):
    fresh = []
    ingredients = None

    for line in input.splitlines():
        if line == '':
            ingredients = []
        elif ingredients is None:
            a, b = line.split('-')
            fresh.append((int(a), int(b)))
        else:
            ingredients.append(int(line))

    return fresh, ingredients

parse(sample)
```

```python
def part1(input):
    fresh, ingredients = input
    result = 0

    for i in ingredients:
        for a, b in fresh:
            if i >=a and i <=b:
                result += 1
                break

    return result

assert(part1(parse(sample)) == 3)
```
