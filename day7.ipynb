```python
sample = '''.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
...............'''
```

```python
def parse(input):
	return input.splitlines()

parse(sample)
```

```python
def part1(input):
	result = 0
	beams = [input[0].find('S')]

	for i in range(1, len(input)):
		new_beams = []
		for j in beams:
			if input[i][j] == '^':
				result += 1
				new_beams += [j-1, j+1]
			else:
				new_beams += [j]
		
		beams = sorted(set(new_beams))
	return result

assert(part1(parse(sample)) == 21)
```

```python
def part2(input):
	mem = {}

	def f(i, j):
		k = (i, j)
		if k not in mem:
			if i == len(input):
				mem[k] = 1
			elif input[i][j] == '^':
				mem[k] = f(i + 1, j - 1) + f(i + 1, j + 1)
			else:
				mem[k] = f(i + 1, j)
		return mem[k]
	
	return f(0, input[0].find('S'))

assert(part2(parse(sample)) == 40)
```
