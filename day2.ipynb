```python
sample = '''11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124'''
```

```python
def parse(input):
    result = []
    for range in input.split(','):
        f, t = range.split('-')
        result.append((int(f), int(t)))
    return result

parse(sample)
```

```python
def is_invalid(id):
    id = str(id)
    mid = len(id) // 2
    return id[:mid] == id[mid:]

assert(is_invalid('55'))
assert(is_invalid('6464'))
assert(is_invalid('123123'))
assert(not is_invalid('101'))

def part1(input):
    result = []
    for (a, b) in input:
        for id in range(a, b + 1):
            if is_invalid(id):
                result.append(id)
    return sum(result)

assert(part1(parse(sample)) == 1227775554)
```

```python
def is_invalid(id):
    id = str(id)
    for l in range(1, (len(id) // 2) + 1):
        if len(id) % l == 0:
            values = [id[i*l:i*l + l] for i in range(len(id) // l)]
            if len(set(values)) == 1:
                return True
    return False

assert(is_invalid('55'))
assert(is_invalid('6464'))
assert(is_invalid('123123'))
assert(is_invalid('123123123'))
assert(not is_invalid('101'))

def part2(input):
    result = []
    for (a, b) in input:
        for id in range(a, b + 1):
            if is_invalid(id):
                result.append(id)
    return sum(result)

assert(part2(parse(sample)) == 4174379265)
```
