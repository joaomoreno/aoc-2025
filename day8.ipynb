```python
from math import sqrt, pow
from functools import reduce

def d(a, b):
	xa,ya,za = a
	xb,yb,zb = b
	return sqrt(pow(xb - xa, 2) + pow(yb - ya, 2) + pow(zb - za, 2))

def part1(input, n=10):
	ds = []

	for i in range(len(input)):
		for j in range(i + 1, len(input)):
			ds.append(((i, j), d(input[i], input[j])))

	ds.sort(key=lambda p : p[1])

	sets = dict()

	for i in range(n):
		((a, b), _) = ds[i]

		if a in sets and b in sets:
			if sets[a] == sets[b]:
				continue

			u = sets[a] | sets[b]
			for x in u:
				sets[x] = u

		elif a in sets:
			sets[a].add(b)
			sets[b] = sets[a]
		elif b in sets:
			sets[b].add(a)
			sets[a] = sets[b]
		else:
			sets[a] = sets[b] = set([a, b])

	sets = set(frozenset(s) for s in sets.values())
	sizes = sorted(list(len(s) for s in sets))
	return reduce(lambda a, b : a*b, sizes[-3:])

assert(part1(parse(sample)) == 40)
```
